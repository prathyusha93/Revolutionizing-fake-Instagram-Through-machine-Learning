# Data Modelling

```python
# Training and testing dataset (inputs)
X_train = instagram_df_train.drop(columns = ['fake'])
X_test = instagram_df_test.drop(columns = ['fake'])
X_train
```

```python
# Training and testing dataset (Outputs)
y_train = instagram_df_train['fake']
y_test = instagram_df_test['fake']
y_train
```

```python
# Scale the data before training the model
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_x = StandardScaler()
X_train = scaler_x.fit_transform(X_train)
X_test = scaler_x.transform(X_test)
```

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes = 2)
y_test = tf.keras.utils.to_categorical(y_test, num_classes = 2)
```

```python
y_train
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(50, input_dim=11, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(150, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(25, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(2,activation='softmax'))

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 50,  verbose = 1, validation_split = 0.1)
```

# Model Validation and Results

```python
print(epochs_hist.history.keys())
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])

plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
plt.show()
```

```python
predicted = model.predict(X_test)
```

```python
predicted_value = []
test = []
for i in predicted:
    predicted_value.append(np.argmax(i))
    
for i in y_test:
    test.append(np.argmax(i))
```

```python
print(classification_report(test, predicted_value))
```

```python
plt.figure(figsize=(10, 10))
cm=confusion_matrix(test, predicted_value)
sns.heatmap(cm, annot=True)
plt.show()
```

# Prediction with images

```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
image_path = r'D:\fake instagram\pic2.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Initialize profile_pic variable
profile_pic = 0  # Default value

# Check if any faces were detected
if len(faces) > 0:
    print(f"Found {len(faces)} face(s) in the image.")
    profile_pic = 1  # Set to 1 if faces are detected
    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
else:
    print("No faces found in the image.")

# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with detected faces using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()

# Output the value of profile_pic
print("Profile pic:", profile_pic)
```

```python
import easyocr
import cv2
import matplotlib.pyplot as plt

# Create an EasyOCR Reader instance for English language
reader = easyocr.Reader(['en'])

# Load the image
image_path = r'D:\fake instagram\pic2.jpg'
image = cv2.imread(image_path)

# Perform text extraction
results = reader.readtext(image)

# Print the results
for (bbox, text, prob) in results:
    print(f"Detected text: {text} (Confidence: {prob:.2f})")
    
    # Draw bounding box around the text
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)

# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with detected text bounding boxes
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
results
```

```python
# Extract text into an array
text_array = [item[1] for item in results]

# Display the extracted text array
print(text_array)
```

```python
# Initialize variables
username = None
posts = None
followers = None
following = None
fullname = None
description = None

# Extract the username
for text in text_array:
    # Check if the text is a valid username (string that is an identifier and not just digits)
    if text.isidentifier() and not text.isdigit():
        username = text
        break

# Extract posts, followers, following, fullname, and description after the username
if username:
    # Find the index of the username in the array
    index = text_array.index(username)

    # Extract the three following numbers (posts, followers, following)
    try:
        posts = text_array[index + 1]
        followers = text_array[index + 2]
        following = text_array[index + 3]
        
        # Skip the three elements (e.g., "Posts", "Followers", "Following") and get the full name
        fullname = text_array[index + 7]  # Full name is the fourth element after these numbers

        # The description is everything after the full name
        description = text_array[index + 8:]  # Get all elements starting after the fullname
    except IndexError:
        pass  # In case the array is shorter than expected

# Display the extracted information
print(f"Username: {username}")
print(f"Posts: {posts}")
print(f"Followers: {followers}")
print(f"Following: {following}")
print(f"Full Name: {fullname}")
print(f"Description: {description}")
```
